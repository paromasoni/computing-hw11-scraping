```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("https://rocktumbler.com/blog/rock-and-mineral-clubs/")
doc = BeautifulSoup(response.content, 'html.parser')
doc.prettify()
```

```python
full = doc.find_all('tr', bgcolor="#FFFFFF")
    # Janie said to specify which 'tr' we want since there are many, and we need this for the parent element eventually
#print (full)
full_list = []
for test in full:
    #print (test)
    name = (test.find_all('td')[0].text)
    city = (test.find_all('td')[1].text)
    linksfull = test.find_all('a')
#     linksfull = test.find_all('a')[0]['href']
#     print (linksfull)
# This by itself gives a list indices must be integers error, not sure why, need to save it as a variable then run
    links = (linksfull[0]['href'])
    parent = (test.parent.parent)
    state = (parent.find_all('h3')[0].text)
    
    eachdict = {'Name': name,
               'Link': links,
               'City': city,
               'State': state}
    
    full_list.append(eachdict)

print (full_list)
    
    
```

```python
import pandas as pd
df = pd.DataFrame (full_list)
df.State = df.State.str.extract(r'^([\w\s]+)Rock')
df.to_csv("rock-clubs.csv")
df
```
