````python
from bs4 import BeautifulSoup
import requests
import pandas as pd
````

````python
response = requests.get("http://www.mineral.k12.nv.us/pages/School_Board_Minutes")
doc = BeautifulSoup(response.content, 'html.parser')
doc.prettify()
````

````python
full = doc.find_all('a')
minutes_list = []
try:
    for links in full:
        if '.pdf' in links['href']:
            url = links['href']
            date = links.text
            minutes = {'Date': date,
                       'URL': url}
            minutes_list.append(minutes)
except:
    pass

print (minutes_list)
    
#     link = (links['href'])
#     if '.pdf' in link:
#         url = link
#         print (url)
   #print (link.text)
        #date = link.text
        #print (url)
````

````python
df = pd.DataFrame (minutes_list)

try:
    df.Date = pd.to_datetime(df.Date)
    df.Date = df.Date.dt.strftime('%Y-%m-%d')
except:
    pass

df.to_csv("minutes.csv")
df
# df['month'] = pd.to_datetime (df['Full-flowering date'], format='%m%d')
# df['month'] = df.month.dt.strftime('%B')
# YYYY-MM-DD
````
